```python
import pandas as pd
```

```python
#Chamando o  arquivo xlsx

filename = r"E:\Codigo_Eleicao\Votacao_RJ.xlsx"
df_eleicaoERJ = pd.read_excel(filename)
                    
df_eleicaoERJ
```

```python
#Filtrando a coluna Município para apenas Niterói

df_eleicaoERJ_Nit = df_eleicaoERJ[df_eleicaoERJ['Município'] == 'NITERÓI']
df_eleicaoERJ_Nit
```

```python
#Criando coluna chamada 'endereco_completo' e colocando dados dentro dessa coluna.

#Defino que quero na nova coluna tenha os dados da coluna 'Endereco do Local', depois ", ", depois os dados da coluna de Bairro e dps ", Niterói"

df_eleicaoERJ_Nit['endereco_completo'] = df_eleicaoERJ_Nit['Endereço do Local'] + ", " + df_eleicaoERJ_Nit['Bairro'] + ", Niterói"
df_eleicaoERJ_Nit
```

```python
!pip install geopandas
```

```python
import geopandas as gpd
```

```python
from geopandas.tools import geocode #Ferramenta de geocodificação para obter as coordenadas de cada endereco
```

```python
#Geocodificando

primeiro_geocode = geocode(df_eleicaoERJ_Nit["endereco_completo"], #Geocodificação a partir da coluna de endereço completo
                provider = "nominatim", #Indicando que a Geocidificação é a partir do OSM
                timeout = None, #Indicando para que nao haja timeout dependendo do tempo de geodificação
                user_agent = "my_application") #identificador
```

```python
primeiro_geocode
```

**Agora vou juntar a tabela geocodificada com a tabela inicial**

```python
#Fazendo um join através dos indices da tabela do excel com a primeira tabela geocodificada

combinacao = primeiro_geocode.merge(df_eleicaoERJ_Nit, #merge da primeira tabela com a geocodificada
                                    left_index = True, #Aqui estou dizendo que no merge é para unir o índice da tabela da esquerda 
                                    right_index = True) #E aqui estou dizendo que a uniao tem que ser com o indice da tabela da direita. Basicamente os indices sao as chaves primarias.
```

```python
#Vendo se as tabelas foram unidas corretamente

combinacao.head()
```

**As tabelas foram unidas corretamente, agora vamos ver quantas linhas não foram geocodificadas**

```python
#Vendo quantas linhas não foram geocodificadas

comb_geom_empty = combinacao[combinacao['geometry'].is_empty]
comb_geom_empty.shape
```

```python
#Vendo a tabela completa somente com as linhas sem geometria

comb_geom_empty.head(10)
```

**Até aqui vimos que de 136 linhas totais, 42 não foram geocodificadas. O que farei a seguir é tentar geocodificar novamente, mas criando**
**uma coluna nova e usando a coluna do local de votação para preencher essa coluna**

```python
#Aqui estou criando uma nova coluna com o local da votação para ver se a geocodificação retorna melhores resultados para os endereços que ficaram sem geometria.

#Estou criando essa coluna na tabela que tem geometria vazia (geom_empty)

comb_geom_empty['endereco_completo_2'] = comb_geom_empty['Local de Votação'] + ", " + comb_geom_empty['Bairro'] + ", Niterói"

#"Conferindo nova coluna"
comb_geom_empty.head()
```

```python
#Realizando o segundo geocode para a tabela de geometrias vazias

segundo_geocode = geocode(comb_geom_empty["endereco_completo_2"], #Geocodificação a partir da coluna de endereço completo 2
                provider = "nominatim", #Indicando que a Geocidificação é a partir do OSM
                timeout = None, #Indicando para que nao haja timeout dependendo do tempo de geodificação
                user_agent = "my_application") #identificador
```

```python
#Vendo quantas linhas não foram geocodificadas

comb_geom_empty_2 = segundo_geocode[segundo_geocode['geometry'].is_empty]
comb_geom_empty_2.shape
```

**Até aqui vemos que das 42 geometrias não codificadas anteriormente, 15 foram encontradas.**

```python
#Visualizando a tabela gerada do segundo geocode

segundo_geocode.head()
```

**Agora podemos juntar a tabela comb_geom_empty (primeira tabelona com geometria vazia e coluna nova) com a tabela do segundo geocode e remover as geometrias vazias**

```python
#Removendo a geometria antiga da primeira combinação para nao ficar com geometria duplicada

comb_geom_empty = comb_geom_empty.iloc[:,1:]
comb_geom_empty.head(10)
```

```python
#Pegando a tabela do segundo geocode e unindo com a tabela de geometrias vazias. Ou seja, ainda temos 42 linhas, porém com algumas agora preenchidas.

segundo_comb_geom = segundo_geocode.merge(comb_geom_empty, left_index = True, right_index = True)
segundo_comb_geom.head()
```

```python
#Nessa nova tabela de remanescentes, estou selecionando as linhas onde a coluna geometry NÃO está vazia. O "~" significa NÃO.

segundo_comb_geom = segundo_comb_geom[~segundo_comb_geom["geometry"].is_empty]

#Como eu disse anteriormente, apenas 15 geometrias foram encontradas.

segundo_comb_geom.shape
```

```python
#Agora estou selecionando as linhas onde a coluna geometry NÃO está vazia na primeira tabela de combinação (onde tem todas geometrias, vazias e nao vazias). O "~" significa NÃO.

combinacao = combinacao [~combinacao ["geometry"].is_empty]
```

```python
#Verificando quantidade de linhas agora na primeira tabela combinada

combinacao.shape
```

```python
combinacao
```

```python
#O método concat permite concatenar DataFrames ou GeoDataFrames ao longo de um eixo específico. 
#O parâmetro ignore_index=True garante que os índices sejam redefinidos após a concatenação.

combinacao_final = pd.concat([combinacao, segundo_comb_geom], ignore_index=True)
```

```python
combinacao_final
```

```python
#Função para saber a quantidade de zonas eleitorais únicas na tabela

len(pd.unique(combinacao_final['Zona Eleitoral']))
```

**Com os dados geocodificados, agora vamos criar um mapa de Voronoi, determinando influencias de acordo com a coluna de preferência**
**Nesse estudo vamos definir as geometrias de influência de cada zona eleitoral. Em Niterói são 4 zonas.**

```python
#Lendo geometria do municipio de niteroi

limites = gpd.read_file(r"E:\RJ_Municipios_2022\RJ_Municipios_2022.shp")
limites
```

```python
#Chamando o municipio de Niteroi

limites = limites[limites['NM_MUN'] == 'Niterói']
limites
```

```python
#Instalando o pacote que gera os polígonos de Voronoi

!pip install geoVoronoi -q
```

```python
#Função que permite gerar poligonos de Voronoi através de coordenadas

from geovoronoi import voronoi_regions_from_coords
```

```python
#Aqui estou selecionando SOMENTE a geometria do limite de Niterói

#iloc[0]: Seleciona a primeira linha do DataFrame limites.
#.geometry: Acessa a coluna "geometry" dessa linha.

area_shape = limites.iloc[0].geometry
```

```python
area_shape
```

```python
#Selecionando todas os dados de local de votacao (pontos) que estão dentro da geometria de Niterói. 
#Isso é importante para não dar erro na hora do mapa de Voronoi. 

comb_sel = combinacao_final[combinacao_final.within(area_shape) == True]
comb_sel

#Ao terminar a analise com within, vimos que todos os pontos de locais de votação estão dentro do polígono de Niterói. Então continua com 109 linhas.
```

```python
comb_sel.plot()
```

**Com os dados de pontos e geometria em mãos, agora vamos criar  o polígono de voronoi**

```python
#Gerando os poligonos de Voronoi

region_polys, region_pts = voronoi_regions_from_coords(comb_sel.geometry, area_shape)
```

```python
#Quando plotamos os poligonos, vemos que isso vem em forma de dicionário, então teremos que fazer uma transformação para dataframe

region_polys
```

```python
df_voronoi = pd.DataFrame.from_dict(region_polys, #Estou pegando todos os objetos do dicionario e to criando um dataframe do tipo pandas baseado no dicionario.
                                    orient = "index").reset_index() #Pra cada item do dicionario estou criando uma nova linha. Depois reseto o indice do dicionario para ele ser numero de 1 até o numero de elementos

df_voronoi
```

```python
#Tranformando nosso DF em GDF

voronoi = gpd.GeoDataFrame(df_voronoi, #Transformando o dataframe em geodataframe
                           geometry = df_voronoi[0], #Especificando a coluna que contém a geometria
                           crs = "EPSG:4326") #Definindo o epsg 

voronoi
```

```python
#Selecionando a coluna de geometria (Selecionando a coluna que ta na segunda posiçao começando do 0)

voronoi = voronoi.iloc[:, 2:3]
voronoi
```

```python
#Esse não é o mapa de influencia das zonas, é um mapa de influencia de cada local de votação

voronoi.plot()
```

```python
#Spatial join para se os pontos dos locais de votação estiverem dentro dos poligonos de voronoi, os atributos deles vao ser combinados aos poligonos de voronoi

eleicao22 = voronoi.sjoin(comb_sel)
```

```python
#Agora nao temos só os polígonos de voronoi, mas todo o resto dos atributos

eleicao22
```

```python
#Agora que nós temos os poligonos de voronoi junto com as zonas de votaçao, podemos agrupar ou dissolver esses poligonos de acordo com as zonas

zonas = eleicao22.dissolve(by = "Zona Eleitoral", 
                           as_index = False) #Nao quero que a zona vire o indice da tabela

zonas.shape #Podemos ver que continua tendo 4 zonas eleitorais
```

```python
#Plot basico dos resultados

zonas.plot(figsize = (10,10),
           column = "Zona Eleitoral",
           legend = True,
           scheme= "quantiles")
```

```python
#Melhorando o plot

import matplotlib.pyplot as plt

# Escolhendo a paleta de cores desejada para compor o mapa. (https://matplotlib.org/stable/users/explain/colors/colormaps.html)
cmap = 'Set3'

#Configurando o mapa em si
ax = zonas.plot(figsize=(10, 10), #Tamanho do mapa
                column="Zona Eleitoral", #Coluna em que ele vai se baseae
                legend=True, #Adicionando legenda
                scheme="quantiles", #Tipo classe do mapa
                cmap=cmap, #Adicionando a cor configurada anteriormente
                edgecolor = 'black', #Cor do contorno do mapa
                linewidth=0.5) #Espessura da linha do mapa

# Defina manualmente os rótulos da legenda e as cores correspondentes 
labels = ['71º Zona Eleitoral', '72º Zona Eleitoral', '144º Zona Eleitoral', '199º Zona Eleitoral'] #Criando uma lista chamada labels contendo os rótulos que serão usados na legenda. 

colors = ['lightgreen', 'orange', 'pink', 'yellow'] #Criando uma lista chamada colors contendo as cores associadas a cada zona eleitoral. (https://matplotlib.org/2.0.2/examples/color/named_colors.html)

#Esta linha cria uma lista de objetos Line2D (linha) que serão usados para criar a legenda.
handles = [plt.Line2D([0], [0], #Cria uma instância da classe Line2D do Matplotlib, que representa uma linha no gráfico. [0], [0]: São as coordenadas x e y da linha. Neste caso, a linha é apenas um marcador, então essas coordenadas não importam muito.
                      marker='s', #Especifica que o marcador usado será um quadrado ('s').
                      color='w', #Define a cor da linha como branca ('w').
                      markerfacecolor=color, # Define a cor do marcador como a cor especificada no loop (color).
                      markersize=10) for color in colors] #Define o tamanho do marcador como 10.

# Atualizando a legenda
ax.legend(handles, #É a lista de objetos Line2D que foram criados para representar as entradas da legenda. Cada objeto contém um marcador quadrado colorido associado a uma zona eleitoral.
          labels, #É a lista de rótulos que serão atribuídos a cada entrada na legenda. Cada rótulo representa uma zona eleitoral.
          title="Zonas Eleitorais", #Define um título para a legenda, que neste caso é "Zonas Eleitorais".
          loc='lower right') #Especifica a posição da legenda no gráfico. Neste caso, a legenda é colocada no canto inferior direito.

# Definindo a cor de fundo
ax.set_facecolor('lightcyan') #https://matplotlib.org/2.0.2/examples/color/named_colors.html

# Adicionando título
plt.title("Mapa de influência por zona eleitoral - Niterói (RJ)")

#Plotando o mapa
plt.show()
```
