```python
import geopandas as gpd
import numpy as np 
from geopandas.tools import geocode #Ferramenta de geocodificação para obter as coordenadas de cada endereco


```

```python
#Criando coluna chamada 'endereco_completo' e colocando dados dentro dessa coluna.

#Defino que quero na nova coluna tenha os dados da coluna 'Endereco do Local', depois ", ", depois os dados da coluna de Bairro e dps ", São Gonçalo"

endereco['endereco_completo'] = endereco['Endereco do Local'] + ", " + endereco['Bairro'] + ", São Gonçalo"
endereco 

```

```python
#Geocodificando

ponto = geocode(endereco["endereco_completo"], #Geocodificação a partir da coluna de endereço completo
                provider = "nominatim", #Indicando que a Geocidificação é a partir do OSM
                timeout = None, #Indicando para que nao haja timeout dependendo do tempom de geodificação
                user_agent = "my_application") #identificador
```

```python
ponto
```

```python
comb = ponto.merge(endereco, left_index = True, right_index = True) #Fazendo um merge através dos indices da tabela do csv com a tabela de pontos gerados
```

```python
comb
```

```python
rem = comb[comb['geometry'].is_empty] #criando variavel para linhas que ficaram sem ponto
rem['endereco_completo'] = rem['Local de Votacao'] + ", " + endereco['Bairro'] + ", São Gonçalo" #criando novos "endereços completos" usando a coluna de local de votaçao para ver se consegue geocodificar
rem
```

```python
rem_ponto = geocode(rem["endereco_completo"], #Geocodificação a partir da coluna de endereço completo
                provider = "nominatim", #Indicando que a Geocidificação é a partir do OSM
                timeout = None, #Indicando para que nao haja timeout dependendo do tempom de geodificação
                user_agent = "my_application") #identificador
```

```python
rem_ponto
```

```python
rem = rem.iloc[:,1:]
rem
```

```python
comb_rem = rem_ponto.merge(rem, left_index = True, right_index = True)
comb_rem.head()
```

```python
comb_rem = comb_rem[~comb_rem["geometry"].is_empty]
```

```python
comb= comb_rem[~comb_rem["geometry"].is_empty]
```
